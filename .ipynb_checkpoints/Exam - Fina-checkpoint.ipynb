## Network #1 - Multi-Layer Perceptron

### Setting variables

```python
# Make a new variable to use on this model from the dataset
mlp_X_train = X_train
mlp_X_test = X_test
mlp_X_valid = X_valid

mlp_y_train = y_train
mlp_y_test = y_test
mlp_y_valid = y_valid

mlp_y_train = np.array(mlp_y_train).reshape(-1,1)
mlp_y_test = np.array(mlp_y_test).reshape(-1,1)
mlp_y_valid = np.array(mlp_y_valid).reshape(-1,1)

### Creating and setting the log directory
# Print the shapes of the resulting sets
print("Train set shape:", mlp_X_train.shape, mlp_y_train.shape)
print("Validation set shape:", mlp_X_valid.shape, mlp_y_valid.shape)
print("Test set shape:", mlp_X_test.shape, mlp_y_test.shape)
```

### Creating and setting the log directory

```python
# Define the root log directory.
mlp_logdir = os.path.join(os.curdir, "adult_data_mlp_logs")
```

```python
# Create a function that will generate a subdirectory path based on the current date and time
# (this information must be passed to the TensorBoard callback as an argument)
def get_run_mlp_logdir():
    import time
    run_id = time.strftime("run_%Y_%m_%d-%H_%M_%S")
    return os.path.join(mlp_logdir, run_id)

run_mlp_logdir = get_run_mlp_logdir()
run_mlp_logdir
```

### Defining callbacks

```python
# Define callbacks.

# ModelCheckpoint.
checkpoint_cb = keras.callbacks.ModelCheckpoint("network2_adult_data", save_best_only=True)

# EarlyStopping (with rollback to the best model).
early_stopping_cb = keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True)

#TensorBoard.
tensorboard_cb = keras.callbacks.TensorBoard(run_mlp_logdir)
```

### Creating the model

```python
# Create a function that builds, compiles and returns a Keras model.
# Note that the function takes a HyperParameters object (hp) as a parameter, which it can
# use to define hyperparameters along with their range of possible values.
import keras_tuner as kt

def build_model(hp):
    n_hidden = hp.Int("n_hidden", min_value=4, max_value=8, default=4)
    n_neurons = hp.Int("n_neurons", min_value=31, max_value=31)
    learning_rate = hp.Float("learning_rate", min_value=200, max_value=300,
                             sampling="log")
    optimizer = tf.keras.optimizers.SGD(learning_rate=learning_rate)
    mlp_model = tf.keras.Sequential()
    mlp_model.add(tf.keras.layers.Flatten())
    for _ in range(n_hidden):
        mlp_model.add(tf.keras.layers.Dense(n_neurons, activation="relu"))
    mlp_model.add(tf.keras.layers.Dense(20, activation="sigmoid"))
    
    
    mlp_model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    
    
    
    return mlp_model
```

### Tuning the hyperparameters

```python
# Use KerasTuner RandomSearch to tune hyperparameters
random_search_tuner = kt.RandomSearch(
    build_model, objective="val_accuracy", max_trials=10, overwrite=True,
    directory="my_mnist", project_name="my_rnd_search", seed=42)

random_search_tuner.search(mlp_X_train, mlp_y_train, epochs=20,
                           validation_data=(mlp_X_valid, mlp_y_valid),
                           callbacks=[checkpoint_cb, early_stopping_cb, tensorboard_cb])
```

```python
# Display a summary of the best model:
best_trial = random_search_tuner.oracle.get_best_trials(num_trials=1)[0]
best_trial.summary()
```

### Training the model

```python
# Get the best hyperparameters from the search
best_hps = random_search_tuner.get_best_hyperparameters(num_trials = 1)[0]

# Build the model with the best hyperparameters
mlp_model = build_model(best_hps)

# Train the model
mlp_history = mlp_model.fit(mlp_X_train, mlp_y_train, epochs=20, validation_data=(mlp_X_valid, mlp_y_valid))
```

### Evaluating the model

```python
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
import matplotlib.pyplot as plt

# Plot the learning curves
plt.plot(mlp_history.history['accuracy'], label='Training accuracy')
plt.plot(mlp_history.history['val_accuracy'], label='Validation accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
# Evaluate the model's accuracy on the test set:
best_model = random_search_tuner.get_best_models(num_models=1)[0]
test_loss, test_accuracy = best_model.evaluate(mlp_X_test, mlp_y_test)
```

```python
# Applying threshold
mlp_y_pred = best_model.predict(mlp_X_test)
mlp_y_pred = (mlp_y_pred > 0.30)
```

### Tuning the hyperparameters

This code performs hyperparameter tuning for a random forest classifier using RandomizedSearchCV.

First, the random forest classifier is defined with a random state of 42. The hyperparameters to search over are defined in a dictionary called param_dist, which includes the number of estimators, maximum depth, minimum samples for splitting, minimum samples for each leaf, and maximum number of features. The number of threads to use is defined as 2, and a RandomizedSearchCV object is defined with 10 iterations, 5-fold cross-validation, and a random state of 42.

The try block uses multithreading to perform the hyperparameter tuning on the training data. It runs the RandomizedSearchCV object for each of the 10 iterations, appending the best parameters found by the RandomizedSearchCV object to a list called results. The tqdm function is used to display a progress bar that updates for each iteration. Finally, the best hyperparameters found are printed for each iteration. If an exception occurs, the error message is printed and the program exits.

```python
# Define the random forest classifier
rf_clf = RandomForestClassifier(random_state=42)

# Define the hyperparameters to search over
param_dist = {
    'n_estimators': np.arange(100, 1001, 100),
    'max_depth': np.arange(10, 101, 10),
    'min_samples_split': np.arange(2, 11),
    'min_samples_leaf': np.arange(1, 11),
    'max_features': ['sqrt', 'log2', None]
}

# Define the number of threads to use
n_jobs = 2

# Define the RandomizedSearchCV object
rs_cv = RandomizedSearchCV(rf_clf, param_distributions=param_dist, n_iter=10, cv=5, n_jobs=n_jobs, random_state=42)

# Use multithreading to perform RandomizedSearchCV on the training data and display a progress bar
try:
    results = []
    for i in tqdm(range(10), desc='Hyperparameter tuning', position=0, leave=True):
        rs_cv.fit(rf_X_train, rf_y_train)
        results.append(rs_cv.best_params_)
        print(f"Iteration {i + 1} completed: {rs_cv.best_params_}")
except Exception as e:
    print(e)
    exit()
```

```python
# Get the best hyperparameters found by RandomizedSearchCV
best_params = rs_cv.best_params_

# Train a Random Forest classifier using the best hyperparameters
rf_model = RandomForestClassifier(random_state=42, **best_params)
rf_model.fit(rf_X_train, rf_y_train)
```

### SHAP (SHapley Additive exPlanations)

SHAP (SHapley Additive exPlanations) is a method used to interpret the predictions of machine learning models. It provides a way to explain the output of any machine learning model by computing the importance of each feature in a given prediction.

The SHAP summary plot that you generated shows the feature importances for the random forest model you trained. The plot ranks the features in descending order of their importance, with the most important features at the top. The importance of each feature is represented by a horizontal bar, where the length of the bar indicates the magnitude of the feature's effect on the model output. The color of the bar indicates the direction of the effect: red indicates a positive effect (i.e., the feature value increases the predicted outcome), and blue indicates a negative effect (i.e., the feature value decreases the predicted outcome).

The SHAP summary plot can be used to identify which features are most important for making predictions with your model. This information can be used to refine your feature selection or to gain insights into the underlying relationships in your data.

```python
import shap

# Define the explainer object
explainer = shap.Explainer(rf_model)

# Compute the SHAP values for each batch of test data using multithreading and display a progress bar
shap_values = []
for i in tqdm(range(0, len(rf_X_test), 1000), desc='Computing SHAP values', position=0, leave=True):
    batch = rf_X_test[i:i+1000]
    if len(batch) < 1000:
        batch = np.vstack((batch, np.zeros((1000 - len(batch), rf_X_test.shape[1]))))
    try:
        shap_values.append(explainer.shap_values(batch))
    except Exception as e:
        print(e)
        continue

# Concatenate the SHAP values for each batch of test data
if shap_values:
    shap_values = np.concatenate(shap_values)
else:
    print('Error: SHAP values list is empty')
```
